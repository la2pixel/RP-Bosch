# Experiment
Control Management Study 1 : **_Is pupil diameter a valid physiological measure in driving intervention studies_?**

The experiment consists of *N*=18 participants being exposed to 8 randomized repeat trials (balanced design) of simulated driving scenarios. There is element of a "driving intervention" within these scenarios which refers to a driver experiencing total loss of control and having no influence over driving trajectory and the automation (Otto) takes over.

**A. AUTHORITATIVE CONTROL INTERVENTIONS**

These interventions occur at some point in between the scenario. The participants are not informed about the intervention occuring during their trial. So the very first time this happens, they won't see this coming. Some participants were fighting the automation, some gave into the intervention and *N*=2 participants experienced "mode confusion"- i.e, did not recognize that the automation has taken over.

Before Otto takes over control, there's a HUD message on the screen which informs participants that an intervention is about to occur. But since the HUD signal proved to be unreliable, we had to exclude this from our analysis. The intervention occurs in 2 different types.

    - Type I: Blocking - Otto blocks user input (T1)
    - Type II: Takeaway - Otto takes control away from the user (T2)

**B. DRIVING SCENARIOS & HMI DESIGN**

Participants have to drive through 4 different driving scenarios in 2 different HMI designs (hence, 8 repeated measures).

    - HMI Design 1: Stronger steering torque resistance
    - HMI Design 2: Softer steering torque resistance

The 4 different driving scenarios (note: (N)SC <= (Non-)safety critical).

    - Scenario 1: Avoiding a Blindspot collision (SC, T1)
    - Scenario 2: Avoiding a wrong turn (NSC, T1)
    - Scenario 3: Avoiding possible skidding (NSC, T2)
    - Scenario 4: Avoiding a head-on collision (SC, T2)

So, in total it would be 18 * 4 * 2= 144 trials

**C. KEY 15 SECONDS OF EVERY SCENARIO**
For the sake of direct comparability between different scenarios, 15 seconds from each scenario was identified and all the 4 scenarios were cropped to 15 s in this manner. The split is:

    - (-5 to 0s): Before intervention
    - (0s to 5s): During intervention 
    - (5s to 10s): After intervention

**D. VARIABLES OF INTEREST**

    - Acceleration [ m/s^2 ] : field.pedalStateAccelerator
    - Braking: [ m/s^2 ] : field.pedalStateBrake
    - Steering Angle [degrees] : steerang_deg
    - Pupil Diameter [mm] : field.eyeTracker.filteredPupilDiameter
    - Event Markers:
        - Scenario number: field.conditionCounter
        - Automation start marker: field.hmiBackgroundState

# Z-Transformation + Baseline

- If we want to run parametric tests to compare differences between groupings, we need to get the distribution as close as possible to Normal.

- Approach:
    - Z-transform PD within each subject to reduce inter-individual variability
    - Baseline PD would be the average of "Before Intervention (in total, not trimmed)" part of the experiment but for each VP separately.
    - Convert PD at each timepoint within each epoch (no. of std deviations from baseline) into z-scores.

- Z-score computed in this way is similar to the baseline-free approach. Although, the advantage here is that we can adjust for large inter-individual differences in PD distributions close to normal.

##### S1 Standardization

```python
import numpy as np
import matplotlib.pyplot as plt

# SCENARIO 1
s1 = ['vp4s1d1', 'vp5s1d1', 'vp6s1d1', 'vp7s1d1', 'vp8s1d1', 'vp10s1d1', 'vp11s1d1', 'vp12s1d1', 'vp13s1d1', 'vp15s1d1', 'vp16s1d1', 'vp17s1d1', 'vp22s1d1', 'vp23s1d1', 'vp24s1d1', 'vp25s1d1', 'vp27s1d1', 'vp28s1d1', 'vp4s1d2', 'vp5s1d2', 'vp6s1d2', 'vp7s1d2', 'vp8s1d2', 'vp10s1d2', 'vp11s1d2', 'vp12s1d2', 'vp13s1d2', 'vp15s1d2', 'vp16s1d2', 'vp17s1d2', 'vp22s1d2', 'vp23s1d2', 'vp24s1d2', 'vp25s1d2', 'vp27s1d2', 'vp28s1d2']

for df_name in s1:
    df = globals()[df_name]
    pd_data = df['field.eyeTracker.filteredPupilDiameter']
    baseline = pd_data.mean()
    std_dev = pd_data.std()
    
    df['Baseline'] = baseline
    df['StdDev'] = std_dev

    # print("Baseline for", df_name, ":", baseline)
    # print("Standard Deviation for", df_name, ":", std_dev)

z_scores = []

# Get the maximum length of time series data
max_length = max(len(globals()[df_name]['ser_rostime_0s']) for df_name in s1)
print("Maximum length of time series data:", max_length)

# Iterate through each participant dataframe
for df_name in s1:
    df = globals()[df_name]
    time_data = df['ser_rostime_0s']
    pd_data = df['field.eyeTracker.filteredPupilDiameter']
    
    # Print available column names for reference
    print("Available columns in dataframe:", df.columns)
    
    # Identify baseline and standard deviation column names
    baseline_column = 'Baseline'  # Replace 'Baseline' with the actual column name
    std_dev_column = 'StdDev'  # Replace 'StdDev' with the actual column name
    
    # Check if baseline and standard deviation columns are present
    if baseline_column not in df.columns or std_dev_column not in df.columns:
        print("Baseline or StdDev columns not found in dataframe", df_name)
        continue
    
    # Identify baseline and standard deviation
    baseline = df[baseline_column].mean()
    std_dev = df[std_dev_column].mean()
    print("Baseline:", baseline)
    print("Standard Deviation:", std_dev)
    
    # Calculate z-scores for each participant
    z_score = (pd_data - baseline) / std_dev
    print("Z-scores for", df_name, ":", z_score)
    
    # Pad z-scores array with NaN values to match the length of the longest array
    padded_z_score = np.pad(z_score, (0, max_length - len(z_score)), mode='constant', constant_values=np.nan)
    
    # Append padded z-scores array to the list
    z_scores.append(padded_z_score)

# Convert the list of padded z-scores arrays to a numpy array
z_scores_concatenated = np.vstack(z_scores)
# print("Shape of concatenated z-scores array:", z_scores_concatenated.shape)
# print("Concatenated z-scores array:", z_scores_concatenated)

mean_z_scores = np.nanmean(z_scores_concatenated, axis=0)
sem_z_scores = np.nanstd(z_scores_concatenated, axis=0) / np.sqrt(len(s1))
time_data_truncated = np.linspace(-5, 10, max_length)

fig, ax = plt.subplots(figsize=(10, 6))
for z_score in z_scores:
    ax.plot(time_data_truncated, z_score, color='lightgrey', zorder=1)

# Plot representative lineplot with 95% CI with higher z-order
ax.plot(time_data_truncated, mean_z_scores, color='blue', label='Representative Pupil Diameter', zorder=2)
ax.fill_between(time_data_truncated, mean_z_scores - sem_z_scores * 1.96, mean_z_scores + sem_z_scores * 1.96, color='orange', alpha=0.3, label='95% CI', zorder=3)
ax.set_xlim(-5, 10)
ax.set_xlabel('Time (in sec)')
ax.set_ylabel('Standardized Pupil Diameter [mm]')
ax.set_title('Pupil Diameter in Scenario 1')
ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

##### SC vs NSC

NULL HYPOTHESIS:  the distributions of pupil diameter values in SC vs NSC are not significantly different.

```python
from scipy.stats import mannwhitneyu

# NON PARAMETRIC Mann-Whitney U test
safety_critical = pup[pup['Scenario'].isin([1, 4])]
non_safety_critical = pup[pup['Scenario'].isin([2, 3])]
statistic, p_value = mannwhitneyu(safety_critical['Totmeddia'], non_safety_critical['Totmeddia'], alternative='two-sided')
print("Mann-Whitney U Test Results:")
print(f"Statistic: {statistic}")
print(f"P-value: {p_value}")

#PARAMETRIC independent samples t-test
from scipy.stats import ttest_ind
safety_critical = pup[pup['Scenario'].isin([1, 4])]['Totmeddia']
non_safety_critical = pup[pup['Scenario'].isin([2, 3])]['Totmeddia']
t_statistic, p_value = ttest_ind(safety_critical, non_safety_critical)
print("Independent Samples t-test Results:")
print(f"t-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```
